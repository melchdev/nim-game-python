## Game Rules

- The game consists of several piles of objects
- Two players take turns
- On each turn, a player removes one or more objects from a single pile
- A player must remove at least one object
- The player who removes the last object wins

## Game Design Decisions

- The game will be implemented as a human vs human game in its initial version
- The initial state of the game will be defined by a fixed list of piles
- The game will run in a command-line style interaction using input/output

## Game Flow

The game loop will follow these steps:

1. Initialize the game state (piles)
2. Set the starting player
3. Display the current state of the game
4. Ask the current player for a move
5. Validate the move
6. Apply the move to the game state
7. Check if the game has ended
8. If the game is not over, switch player and repeat
